```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.utils.multiclass import unique_labels
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# 1.3Ghz

```python
dataset=pd.read_csv("For_1.3Ghz.csv")
dataset
```

```python
#dataset.dtypes
```

```python
le = preprocessing.LabelEncoder()
```

```python
X=dataset[['CPU LOAD', 'MEMORY USAGE', 'PAGE FAULT', 'CPU IDLE TIME', 'CACHE MISS', 'TEMPERATURE']]
y=dataset['LABEL']
```

```python
y=y.astype('int') 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_test
```

```python
y_train
```

```python
sc=StandardScaler()
```

```python
scaler = sc.fit(X_train)
```

```python
x_train_scaled = scaler.transform(X_train)
```

```python
x_test_scaled = scaler.transform(X_test)
```

```python
#clf = MLPClassifier(random_state=0, max_iter=100).fit(x_train_scaled, y_train)
```

```python
#clf = MLPClassifier(hidden_layer_sizes=(4,4),
                  # random_state=5,
                  # verbose=True,
                   # activation = 'relu',
                    #solver = 'sgd',
                    #learning_rate_init=0.01)
clf = MLPClassifier(solver='sgd',verbose=True,
                    activation = 'relu',
                    hidden_layer_sizes=(4,4), 
                    random_state=1)
```

```python
clf.fit(x_train_scaled,y_train)
```

```python
y_pred=clf.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# 2Ghz

```python
dataset1=pd.read_csv("For_2Ghz.csv")
dataset1
```

```python
dataset1.dtypes
```

```python
le = preprocessing.LabelEncoder()
```

```python
X=dataset1[['CPU LOAD', 'MEMORY USAGE', 'PAGE FAULT', 'CPU IDLE TIME', 'CACHE MISS', 'TEMPERATURE']]
y=dataset1['LABEL']
```

```python
y=y.astype('int') 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
clf = MLPClassifier(hidden_layer_sizes=(4,4),
                    random_state=0,
                    verbose=True,
                    activation = 'relu',
                    solver = 'sgd',
                    learning_rate_init=0.01)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
